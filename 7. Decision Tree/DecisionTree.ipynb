```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/ML_practise_project/salaries.csv")
df.head()
```

```python
input = df.drop("salary_more_then_100k", axis="columns")
target = df["salary_more_then_100k"]
input
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
l_company = LabelEncoder()
l_job = LabelEncoder()
l_degree = LabelEncoder()
```

```python
input["company_n"] = l_company.fit_transform(input["company"])
input["job_n"] = l_company.fit_transform(input["job"])
input["degree_n"] = l_company.fit_transform(input["degree"])
input.head()
```

```python
input_n = input.drop(["company", "job", "degree"], axis="columns")
input_n
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(input_n, target)
```

```python
model.score(input_n, target)
```
